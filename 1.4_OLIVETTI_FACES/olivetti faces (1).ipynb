Find and load the Olivetti faces dataset

```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
olivetti_faces = fetch_olivetti_faces(shuffle=True,random_state=42)
```

 Description for this dataset

```python
print("Description of Olivetti faces  dataset:")
print(olivetti_faces.DESCR)
```

Targets

```python
targets = olivetti_faces.target
print("\nTargets:")
print(targets)
```

Features

```python
features = olivetti_faces.data
print("\nFeatures:")
print(features)
```

Data an labeled data

```python
print("\nData shape:", features.shape)
print("Targets shape:", targets.shape)
```

To print Shape and datatype of various items

```python
print("\nTarget Shape:")
print(targets.shape)
print("\nFeatures Datatype:")
print(features.dtype)
```

Display varios items

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))

for i, ax in enumerate(axes.flat):
    ax.imshow(features[i].reshape(64, 64), cmap='gray')
    ax.text(0.05, 0.05, str(targets[i]), transform=ax.transAxes, color='green')

plt.show()
```

plotting the digits 
and labeling the image with the target value

```python
fig, axes = plt.subplots(2,5,figsize=(10,4),
                        subplot_kw={'xticks':[],'yticks':[]},
                        gridspec_kw=dict(hspace=0.1,wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(features[i].reshape(64,64),cmap='gray')
    ax.text(0.05,0.05,str(targets[i]),transform=ax.transAxes,
           color='green' if targets[i] == 1 else 'red') 
plt.show() 
```
