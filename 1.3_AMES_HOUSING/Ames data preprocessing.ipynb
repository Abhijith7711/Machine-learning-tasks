## Handling duplicate values

```python
data = data.drop_duplicates()
```

```python
continuous_columns = data.select_dtypes(include=['int64', 'float64'])
```

```python
continuous_columns_names = continuous_columns.columns.tolist()
```

```python
cor_df = continuous_columns.corr()
```

```python
cor_df['SalePrice'][0]
```

```python
column_index_del = []
for i in range(0,39):
    if (cor_df['SalePrice'][i]>0.50 or cor_df['SalePrice'][i]<-0.15):
        pass
    else:
        column_index_del.append(i)
column_index_del = column_index_del[:-1]    
```

```python
num_col_considered = continuous_columns.drop(continuous_columns.columns[column_index_del], axis=1)
y = data['SalePrice']
X = num_col_considered
X.head()
```

```python
X.info()
```

## Handling missing values

```python
mean_values = round(X.mean(),2)
mean_values
```

```python
X.fillna(mean_values,inplace = True)
```

```python
X.info()
```

```python
col_x_list = X.columns.tolist()
for i in col_x_list:
    X[i] = np.log1p(X[i])
X.head()
```

## Applying Min-Max-Scaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
columns_to_scale = X.columns.tolist()
X[columns_to_scale] = scaler.fit_transform(X[columns_to_scale])
print(X)
```

## Handling Outliers

```python
from scipy import stats
z_scores = stats.zscore(X)
threshold = 3
X = X[(z_scores < threshold).all(axis=1)]
X
```

```python
categorical_columns = data.select_dtypes(include=['object'])
```

## Conversion of categorical data into continous

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
df_encoded = categorical_columns.copy()
lis = categorical_columns.columns.tolist()
df_encoded[lis] = encoder.fit_transform(categorical_columns[lis])
print(df_encoded)
```

```python
col_del = []
for i in range(43):
    if (df_encoded.isnull().sum()[i]>0):
        col_del.append(i)
```

```python
cat_df = df_encoded.drop(df_encoded.columns[col_del], axis=1)
```

```python
cat_df['y'] = y
```

```python
cor_cat_df = cat_df.corr()
```

```python
col_t_considered = []
for i in range(28):
    if(cor_cat_df['y'][i]>0.25 or cor_cat_df['y'][i]<-0.15):
        pass
    else:
        col_t_considered.append(i)
```

```python
col_t_considered = col_t_considered[:-1]
```

```python
cat_cols = cat_df.drop(cat_df.columns[col_t_considered],axis=1)
cat_cols
```

```python
X = pd.merge(X, cat_cols, left_index=True, right_index=True)
```

```python
X.head()
```

```python
tot_data = pd.merge(X,y,left_index=True,right_index=True)
tot_data.head()
```
